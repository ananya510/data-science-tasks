```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
  data = pd.read_csv ("/content/heart.csv")
  data
```

```python
print("Missing values:\n", data.isnull().sum())
```

```python
data["sex"].replace({1: "MALE", 0: "FEMALE"}, inplace =True)
data.head()
```

```python
print("\nNumber of duplicates:", data.duplicated().sum())
```

```python
data = data.drop_duplicates()
```

```python
data.describe(include='all').T
```

```python
data.query("sex == 'MALE'")
```

```python
data.query("sex != 'MALE'")
```

```python
data.query("age > 30 and sex != 'MALE'")
```

```python
categorical_cols = [col for col in data.columns if data[col].nunique() <= 10]
categorical_cols
```

```python
numerical_cols = [col for col in data.columns if data[col].nunique() > 10]
numerical_cols
```

```python
for col in categorical_cols:
    print(f"\nUnique values in {col}: {data[col].unique()}")
```

```python
data[numerical_cols].hist(figsize=(12, 10))
plt.suptitle('Histograms of Numerical Columns', y=0.92)
plt.show()
```

```python
plt.figure(figsize=(12, 10))
for i, col in enumerate(numerical_cols, 1):
    plt.subplot(3, 3, i)
    sns.boxplot(y=data[col])
    plt.title(col)
plt.suptitle('Box Plots of Numerical Columns', y=0.92)
plt.tight_layout()
plt.show()
```

```python
print("\nDescriptive Statistics of Numerical Columns:")
print(data[numerical_cols].describe().to_markdown(numalign="left", stralign="left"))
```

```python
target_counts = data['target'].value_counts()
plt.figure(figsize=(8, 6))
plt.pie(target_counts, labels=[f'Class {i}' for i in target_counts.index], autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Target Variable')
plt.show()
```

```python
for col in categorical_cols:
    if col != 'target':
        cross_tab = pd.crosstab(data[col], data['target'])
        cross_tab.plot(kind='bar', stacked=True, figsize=(8, 6))
        plt.title(f'{col} vs. Target')
        plt.xlabel(col)
        plt.ylabel('Count')
        plt.xticks(rotation=0)
        plt.show()
```

```python
for col in numerical_cols:
    plt.figure(figsize=(8, 6))
    sns.boxplot(x='target', y=col, data=data)
    plt.title(f'{col} vs. Target')
    plt.xlabel('Target')
    plt.ylabel(col)
    plt.show()
```

```python
corr_matrix = data[numerical_cols].corr()
corr_matrix
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of Numerical Columns')
plt.show()
```
